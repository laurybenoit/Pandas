```python
import pandas as pd
import numpy as np
```

```python
files_import = "Resources/purchase_data.csv"
purchase_data_df = pd.read_csv(files_import)
purchase_data_df.head()
```

```python
#counting the number of players
total_players_df = len(purchase_data_df)
total_players_df = pd.DataFrame({"Total Players":total_players_df}, index = [0])
total_players_df
```

```python
# unique Items 
unique_items_df = purchase_data_df["Item Name"].nunique()
#average price
average_price_df = purchase_data_df["Price"].mean()
#number of purchases
number_of_purchases = purchase_data_df["Purchase ID"].count()
#total rev
total_rev = purchase_data_df["Price"].sum()
summary_table = pd.DataFrame({"Number of Unique Items": [unique_items_df], "Average Price":[average_price_df], "Number of purchases":[number_of_purchases], "Total Revenue":[total_rev] })
summary_table
```

```python
#count gender
gender_count_df = purchase_data_df.groupby ("Gender")["SN"].nunique()
gender_count_df.nunique()
#sum of gender
gender_count_df
```

```python
# Gender percentage
percentage_gender = gender_count_df / gender_count_df.sum()
percentage_gender.round()
percentage_gender.head()
```

```python
summary = pd.DataFrame({"Total Count": gender_count_df, "Percentage":percentage_gender })
summary.head()
```

```python
#Purchasing Analysis (Gender)
# count by gender
purchase_count = gender_count_df.nunique()
purchase_count
average_price = gender_count_df / purchase_count
average_price
sum_price = gender_count_df.mean()
average_person = average_price / average_price.sum()
average_person
purchase_analysis = pd.DataFrame({"Purchase Count":purchase_count, "Average Price":average_price,"Average Per Person":average_person,"Total":sum_price})
purchase_analysis
```

```python
# Age group
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data_df['total count'] = pd.cut(purchase_data_df['Age'],bins, labels= group_names)
purchase_data_df
groups_of_age = purchase_data_df.groupby('total count')
groups_of_age_df = groups_of_age.count()
groups_of_age_df
age_group_only = groups_of_age_df['Age']
age_group_only
final_ages = pd.DataFrame({"Age Groups":age_group_only})
final_ages
```

```python
# Top Spenders
spender = purchase_data_df.groupby("SN")
count_spender = spender["Purchase ID"].count()
count_spender
item_names = spender["Item Name"]
item_names
spender_avg = spender["Price"].mean()
total_spender = spender["Price"].sum()
top_spender = pd.DataFrame({"Purchase Count": count_spender, "Item Name":item_names,"Average Purchase Price":spender_avg,"Total Purchase Value": total_spender})
top_spender.head(10)
```
